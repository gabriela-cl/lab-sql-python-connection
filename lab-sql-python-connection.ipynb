```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine, text
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
sakila = "sakila"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+sakila
engine = create_engine(connection_string)
engine
```

```python
def rental_date():
    with engine.connect() as connection:
        txt = """SELECT rental_date
                FROM rental;"""
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
```

```python
rental_date()
```

```python
def rentals_month(engine, month, year):
    with engine.connect() as connection:
        txt = f"""SELECT *
                 FROM rental
                 WHERE MONTH(rental_date) = {month} AND YEAR(rental_date) = {year};"""
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
```

```python
df_june_2005 = rentals_month(engine, 6, 2005)
df_june_2005
```

```python
df_may_2005 = rentals_month(engine, 5, 2005)
df_may_2005
```

```python
#3. number of rentals made by each customer_id during the selected month and year.

def rental_count_month(rentals_month_df, month, year):
    with engine.connect() as connection:
        txt = f"""SELECT customer_id, COUNT(rental_id) AS rentals_{month}_{year}
                 FROM rental
                 WHERE MONTH(rental_date) = {month} AND YEAR(rental_date) = {year}
                 GROUP BY customer_id;"""
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
```

```python
#3 correction:

def rental_count_month(df, month, year=2005):
    rentals_by_customer = df.loc[(df['rental_date'].dt.month == month) & (df['rental_date'].dt.year == year)]\
                          .groupby('customer_id')['rental_id'].count().reset_index()
    rentals_by_customer = rentals_by_customer.rename(columns={'rental_id': f'rentals_{month}_{year}'})
    return rentals_by_customer
```

```python
#4

def compare_rentals(rentals_month_df, rentals_month_df2):
    with engine.connect() as connection:
        txt = f"""SELECT {rentals_month_df}.customer_id, 
                 ({rentals_month_df}[rentals_{month}_{year}] - {rentals_month_df2}[rentals_{month}_{year}]) AS difference
                 FROM {rentals_month_df}
                 INNER JOIN {rentals_month_df2}
                 USING customer_id
                 GROUP BY customer_id;"""
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
```

```python
def compare_rentals(rentals_month_df, rentals_month_df2):
    with engine.connect() as connection:
        txt = f"""SELECT r1.customer_id, 
                         r1.number_of_rentals - r2.number_of_rentals AS difference
                  FROM (
                      SELECT customer_id, COUNT(rental_id) AS number_of_rentals
                      FROM {rentals_month_df}
                      GROUP BY customer_id
                  ) AS r1
                  INNER JOIN (
                      SELECT customer_id, COUNT(rental_id) AS number_of_rentals
                      FROM {rentals_month_df2}
                      GROUP BY customer_id
                  ) AS r2
                  ON r1.customer_id = r2.customer_id;"""
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
```

```python
compare_rentals(df_may_2005, df_june_2005)
```

```python
# 4 correction:

def compare_rentals(df1, df2):
    merged_df = pd.merge(df1, df2, on='customer_id')
    merged_df['difference'] = merged_df[df1.columns[1]] - merged_df[df2.columns[1]]
    return merged_df
```
